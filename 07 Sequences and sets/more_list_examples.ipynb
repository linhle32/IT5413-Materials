```python
#this program simulates a check-out system at a store
#you can enter each item you bought with its price
#the program will produces the receipt with all items and prices
#and a total

##
#we will have two empty lists, one for items, and one for prices
items = []
prices = []

#we will keep looping until the guest enter 'exit()'
#item is the user input for item name, we will also use it to check what the user typed
item = ''

#this is an infinity loop
#it stops only when user type 'exit()'
while True:
    item = input('Item name (please enter exit() to stop): ')
    
    #check if the user typed 'exit()'
    if item == 'exit()':
        #break will stop the current loop
        break
    
    price = float(input('-- Price: '))
    
    #appending the current item to the end of items list
    items.append(item)
    #appending the current item's price to the end of the prices list
    prices.append(price)

#prepare the receipt
print()
print()
print('Receipt')
print('-------')

for i in range(len(items)):
    #%-10s allows us to print string variables in 10 characters, and align them to the left
    #just to have nicely formatted receipt
    print('%-10s %.2f' % (items[i], prices[i]))
    
print('-------')    
print('%-10s %.2f' % ('Total: ', sum(prices)))
```

```python
# this program simulates a check-out system at a store
#you can enter each item you bought with its price
#the program will produces the receipt with all items and prices
#and a total

##
#we will have two empty lists, one for items, and one for prices
items = []
quantities = []
prices = []

#we will keep looping until the guest enter 'exit()'
#item is the user input for item name, we will also use it to check what the user typed
item = ''
total = 0

#this is an infinity loop
#it stops only when user type 'exit()'
while True:
    item = input('Item name (please enter exit() to stop): ')
    
    #check if the user typed 'exit()'
    if item == 'exit()':
        #break will stop the current loop
        break
    
    quantity = int(input('-- Quantity: '))
    price = float(input('-- Price: '))
    
    #adding current item's total to overall total
    total += quantity*price
    
    #appending the current item to the end of items list
    items.append(item)
    #appending the current item's quantity to the end of quantities list
    quantities.append(quantity)
    #appending the current item's price to the end of the prices list
    prices.append(price)

#prepare the receipt
print()
print()
print('Receipt')
print('-'*32)
print('%-10s %10s %10s' % ('Item','Quantity','Price'))

for i in range(len(items)):
    #%-10s allows us to print string variables in 10 characters, and align them to the left
    #just to have nicely formatted receipt
    print('%-10s %10d %10.2f' % (items[i], quantities[i], prices[i]))
    
print('-'*32)    
print('%-10s %10s %10.2f' % ('Total: ', '', total))
```

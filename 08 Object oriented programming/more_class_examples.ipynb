This is an example extending the problem from your lab 4 - we will write classes for employee and department. 

- Class Employee: has id, first name, last name, and salary, all hidden with setters and getters
- Class Department: has name, and employees which is a list of all employee. Both are hidden, but employees cannot be changed so there is no setter for employees. Right now there is only add employee method, we can add remove employee later. We also have methods to return the number of employees and total salary in the department

The class diagram is as below. Remember, the connection with the diamond means 'aggregation' - one department consists of multiple employees



```python
#this class is simple, just four properties and their getters and setters, and __str__()
class Employee:
    
    def __init__(self,eid,first_name,last_name,salary):
        self.__id = eid
        self.__first_name = first_name
        self.__last_name = last_name
        self.__salary = salary
    
    def get_id(self):
        return self.__id
    
    def get_first_name(self):
        return self.__first_name
        
    def get_last_name(self):
        return self.__last_name
    
    def get_salary(self):
        return self.__salary
    
    def set_id(self, eid):
        self.__id = eid
    
    def set_first_name(self, first_name):
        self.__first_name = first_name
        
    def set_last_name(self, last_name):
        self.__last_name = last_name
    
    def set_salary(self, salary):
        self.__salary = salary
        
    def __str__(self):
        #we need to use str() to convert salary from float to string
        return self.__id + ', ' + self.__first_name + ' ' + self.__last_name + ', ' + str(self.__salary)
```

```python
#testing the class Employee
alice = Employee('000123','Alice','Smith',80000)
bob = Employee('000456','Bob','Johnson',75000)
print(alice)
print(bob)
```

```python
class Department:
    
    def __init__(self,department_name):
        self.__name = department_name
        self.__employees = []
    
    #getters
    def get_name(self):
        return self.__name
    
    def get_all_employees(self):
        return self.__employees
    
    #setter - we won't allow changing employees to a totally new list
    #so it has no setters
    def set_name(self, new_name):
        self.__name = new_name
    
    #this method add a single employee to the current employees list
    def add_employee(self, employee):
        self.__employees.append(employee)
    
    #removing employee will be implemented later
    #
    
    #this method returns the number of employees in the department
    def size(self):
        return len(self.__employees)
    
    #this method returns the total salary of the whole department
    def total_salary(self):
        total = 0
        for employee in self.__employees:
            total += employee.get_salary()
        return total
```

```python
#testing the class Department
it = Department('Information Technology')

it.add_employee(alice)
it.add_employee(bob)
it.add_employee(Employee('000543','Carol','Davenport',90000))
it.add_employee(Employee('001543','Daniel','Williams',85000))
it.add_employee(Employee('007525','Emma','Greene',87500))

print('Department of', it.get_name())
print('Department size: ', it.size())

#print all employees in department
for employee in it.get_all_employees():
    print(employee)
    
print('Total Salary: ', it.total_salary())
```
